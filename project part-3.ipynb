```python
import json

data = json.loads(response.text)

rankings = data['rankings']


rankings_list = []
competitor_ranking_list = []
competitor_list = []


for comp in rankings:
    ranking_id = str(comp["type_id"]) + "_" + str(comp["year"]) + "_" + str(comp["week"]) + "_" + comp["gender"]

    rankings_list.append({
        "ranking_id": ranking_id,
        "type_id": comp["type_id"],
        "name": comp["name"],
        "year": comp["year"],
        "week": comp["week"],
        "gender": comp["gender"]
    })

    
    for cr in comp.get("competitor_rankings", []):
        competitor_ranking_list.append({
            "ranking_id": ranking_id,
            "rank": cr["rank"],
            "movement": cr["movement"],
            "points": cr["points"],
            "competitions_played": cr["competitions_played"],
            "competitor_id": cr["competitor"]["id"]
        })

        
        competitor = cr["competitor"]
        is_existing = False
        for c in competitor_list:
            if c["id"] == competitor["id"]:
                is_existing = True
                break
        
        if not is_existing:
            competitor_list.append({
                "id": competitor["id"],
                "name": competitor["name"],
                "country": competitor.get("country"),
                "country_code": competitor.get("country_code"),
                "abbreviation": competitor.get("abbreviation")
            })


print("Rankings List Example:")
print(rankings_list)

print("\nCompetitor Rankings Example:")
print(competitor_ranking_list)

print("\nCompetitor List Example:")
print(competitor_list)
```

```python
mycursor.execute("""
    CREATE TABLE competitors (
    id VARCHAR(50) PRIMARY KEY,
    name VARCHAR(100) NOT NULL,
    country VARCHAR(100) NOT NULL,
    country_code CHAR(3) NOT NULL,
    abbreviation VARCHAR(10) NOT NULL
)
                 """)
```

```python
from tabulate import tabulate
```

```python
# Get all competitors with their rank and points
mycursor.execute("""
SELECT c.name, cr.rank, cr.points
FROM competitors c
JOIN competitor_rankings cr ON c.id = cr.competitor_id
""")
out=mycursor.fetchall()
print(tabulate(out, headers=[i[0] for i in mycursor.description],  tablefmt='grid'))
```

```python
#Find competitors ranked in the top 5
mycursor.execute("""
SELECT c.name, cr.rank, cr.points
FROM competitors c
JOIN competitor_rankings cr ON c.id = cr.competitor_id
WHERE cr.rank <= 5
ORDER BY cr.rank
""")
out=mycursor.fetchall()
print(tabulate(out, headers=[i[0] for i in mycursor.description],  tablefmt='grid'))
```

```python
#List competitors with no rank movement (stable rank)
mycursor.execute("""
SELECT c.name, cr.rank, cr.movement
FROM competitors c
JOIN competitor_rankings cr ON c.id = cr.competitor_id
WHERE cr.movement = 0
""")
out=mycursor.fetchall()
print(tabulate(out, headers=[i[0] for i in mycursor.description],  tablefmt='grid'))
```

```python
#Get total points of competitors from Croatia
mycursor.execute("""
SELECT SUM(cr.points) AS total_points
FROM competitors c
JOIN competitor_rankings cr ON c.id = cr.competitor_id
WHERE c.country = 'Croatia'
""")
out=mycursor.fetchall()
print(tabulate(out, headers=[i[0] for i in mycursor.description],  tablefmt='grid'))
```

```python
 #Count number of competitors per country
mycursor.execute("""
SELECT country, COUNT(*) AS competitor_count
FROM competitors
GROUP BY country
ORDER BY competitor_count DESC
""")
out=mycursor.fetchall()
print(tabulate(out, headers=[i[0] for i in mycursor.description],  tablefmt='grid'))
```

```python
#Find competitor with the highest points
mycursor.execute("""
SELECT c.name, cr.points
FROM competitors c
JOIN competitor_rankings cr ON c.id = cr.competitor_id
ORDER BY cr.points DESC
LIMIT 1
""")
out=mycursor.fetchall()
print(tabulate(out, headers=[i[0] for i in mycursor.description],  tablefmt='grid'))
```
